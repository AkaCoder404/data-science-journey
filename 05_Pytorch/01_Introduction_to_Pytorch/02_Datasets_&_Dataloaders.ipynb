## Loading a Dataset

```python
import torch
from torchvision import datasets
from torchvision.transforms import ToTensor
import matplotlib.pyplot as plt
```

```python
# Load FashionMNIST dataset

training_data = datasets.FashionMNIST(
    root="~/Developer/Datasets/", # root is the path where the train/test data is stored
    train=True, # train=True if we want to load the training data, else False
    download=True, # download=True if we want to download the data from the internet and put it in the root directory
    transform=ToTensor() # transform is a function that takes in an image and returns a transformed version
)

test_data = datasets.FashionMNIST(
    root="~/Developer/Datasets/",
    train=False,
    download=True,
    transform=ToTensor()
)
```

## Iterating and Visualizing the Dataset

```python
mnist_classes = datasets.FashionMNIST.classes
labels_map = dict(enumerate(mnist_classes))
print(labels_map)

figure = plt.figure(figsize=(8, 8))
cols, rows = 3, 3
for i in range(1, cols*rows+1):
    sample_idx = torch.randint(len(training_data), size=(1,)).item()
    img, label = training_data[sample_idx]
    figure.add_subplot(rows, cols, i)
    plt.title(labels_map[label])
    plt.axis("off")
    plt.imshow(img.squeeze(), cmap="gray")
    
plt.show()
```
