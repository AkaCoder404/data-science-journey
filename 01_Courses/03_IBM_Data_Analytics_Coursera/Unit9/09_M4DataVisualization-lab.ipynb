## Demo: How to run an sql query


```python
# print how many rows are there in the table named 'master'
QUERY = """
SELECT COUNT(*)
FROM master
"""

# the read_sql_query runs the sql query and returns the data as a dataframe
df = pd.read_sql_query(QUERY,conn)
df.head()
```

## Demo: How to list all tables


```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

## Demo: How to run a group by query


```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

## Demo: How to describe a table


```python
table_name = 'master'  # the table you wish to describe

QUERY = """
SELECT sql FROM sqlite_master
WHERE name= '{}'
""".format(table_name)

df = pd.read_sql_query(QUERY,conn)
print(df.iat[0,0])
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
import seaborn as sns
table_name  
QUERY = """
SELECT ConvertedComp FROM master
"""
df = pd.read_sql_query(QUERY,conn)

sns.histplot(df, x="ConvertedComp", bins=100)


```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
QUERY = """
SELECT Age FROM master
"""
df = pd.read_sql(QUERY, conn)
sns.boxplot(x=df["Age"])
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY = """ 
SELECT Age, WorkWeekHrs FROM master
"""

df = pd.read_sql(QUERY, conn)
sns.set(rc={'figure.figsize':(12, 9)})
sns.scatterplot(x="Age", y="WorkWeekHrs", data=df)
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY = """
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
"""
df = pd.read_sql(QUERY,conn)
sns.set(rc={'figure.figsize':(12, 9)})
sns.scatterplot(x="WorkWeekHrs", y="CodeRevHrs", hue="Age", size="Age", data=df, sizes=(20,100))
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
import matplotlib.pyplot as plt
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(DatabaseDesireNextYear) as C FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY C DESC
"""

df = pd.read_sql(QUERY, conn)
print(df.head())

plt.figure(figsize=(10, 10))
# white background
plt.style.use('ggplot')
plt.pie(data=df[0:5], x='C', labels='DatabaseDesireNextYear', autopct='%1.1f%%')
plt.title("Top 5 Databases to learn in 2020")
plt.legend()
plt.show()


```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """ 
SELECT WorkWeekHrs, CodeRevHrs, Age FROM master
WHERE AGE BETWEEN 30 AND 35
GROUP BY Age
"""

df = pd.read_sql(QUERY, conn)
print(df.head())

# df = df.groupby('Age', axis=0).median()
df['WorkWeekHrs'] = df['WorkWeekHrs'].median()
df['CodeRevHrs'] = df['CodeRevHrs'].median()
print(df.head())

df.plot(kind='bar', x='Age', stacked=True, figsize=(20, 10))

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

QUERY = """ 
SELECT ConvertedComp, Age FROM master
WHERE AGE BETWEEN 25 AND 30
"""

df = pd.read_sql(QUERY, conn)
print(df.head())
df = df.groupby('Age').median()

df.plot(kind='line', figsize=(20, 8), color='green')
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY =  """
SELECT MainBranch, COUNT(MainBranch) as C FROM master
GROUP BY MainBranch
"""
df = pd.read_sql(QUERY, conn)

df.plot(kind='barh',x='MainBranch', figsize=(20, 10), color='green', zorder=2, width=0.85)
```

```python
# Database 
QUERY = """
SELECT Respondent, DatabaseWorkedWith, Count(*) AS Count
FROM DatabaseWorkedWith
GROUP BY Respondent
HAVING Count = 1 
   AND DatabaseWorkedWith = 'MySQL'
"""

df = pd.read_sql(QUERY, conn)

print(df.head())
print(df.shape)
```

```python
QUERY = """
SELECT DevType, Count(DevType) FROM DevType
GROUP BY DevType
ORDER BY Count(DevType) DESC
"""

df = pd.read_sql(QUERY, conn)

print(df.head())
print(df.shape)
```

```python
# Top 5 Programming Languages This Year
QUERY = """
SELECT LanguageWorkedWith, COUNT(LanguageWorkedWith) as C FROM LanguageWorkedWith
GROUP BY LanguageWorkedWith
ORDER BY C DESC
"""

df = pd.read_sql(QUERY, conn)
df.head()

df[:5].plot(kind='bar', x='LanguageWorkedWith', y='C', figsize=(8, 10), rot=0)
plt.title("Top 5 Programming Languages This Year")
plt.xlabel("Programming Languages")
plt.ylabel("Number of People Who Use")
```

```python
# Top 5 Database Trends 
QUERY = """
SELECT DatabaseWorkedWith, COUNT(DatabaseWorkedWith) as C FROM DatabaseWorkedWith
GROUP BY DatabaseWorkedWith
ORDER BY C DESC
"""

df = pd.read_sql(QUERY, conn)
df.head()

df[:5].plot(kind='bar', x='DatabaseWorkedWith', y='C', figsize=(8, 10), rot=0)
plt.title("Top 5 Database Worked With This Year")
plt.xlabel("Database")
plt.ylabel("Number of People Who Use")
```

```python
# Top 5 Database Trends for Next Year
QUERY = """
SELECT DatabaseDesireNextYear, COUNT(DatabaseDesireNextYear) as C FROM DatabaseDesireNextYear
GROUP BY DatabaseDesireNextYear
ORDER BY C DESC
"""

df = pd.read_sql(QUERY, conn)
df.head()

df[:5].plot(kind='bar', x='DatabaseDesireNextYear', y='C', figsize=(8, 10), rot=0)
plt.title("Top 5 Database Desired for Next Year")
plt.xlabel("Database")
plt.ylabel("Number of People Who Use")
```
