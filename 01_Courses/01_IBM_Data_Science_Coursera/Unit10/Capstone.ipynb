## Collecting Data with SpaceX REST API
https://api.spacexdata.com/v4/launches/past

```python
# import
import requests
import pandas as pd
```

```python
# get spacex data via api
url = 'https://api.spacexdata.com/v3/launches/past'
response = requests.get(url)
data = response.json()
data_df = pd.json_normalize(data) # normalize json data into a dataframe
print(data_df.shape)
data_df.head()
```
